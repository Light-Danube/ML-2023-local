### 3. Відновлення пропущених значень

на цьому етапі треба відновити `NaN` значення в показчиках, вибравши та застосувавші відповідну стратегію

__Корисні посилання__:
- [Working with missing data](https://pandas.pydata.org/docs/user_guide/missing_data.html)
- [Missing values in pandas (nan, None, pd.NA)](https://note.nkmk.me/en/python-pandas-nan-none-na/)

```python
#З'ясуємо, яка поточна кількість пустих значень у файлі:
apartments.isnull().sum()
```

```python
# заповнити незначащі числові показчики медіаной відповідної колонки
for emp in ['price_per_m2', 'year', 'area_total', 'area_living', 'area_kitchen']:
    apartments[emp] = apartments[emp].fillna(apartments[emp].median())
```

```python
# заповнити незначащі категоріальні показчики вибравши стратегію "найближчий попередній (bfill)"
for emp in ['rooms', 'level', 'levels']:
    apartments[emp] = apartments[emp].fillna(method='bfill')
```

```python
# проінспетувати стан отриманого датасету
print(apartments.isnull().sum())

print('All values are filled, except of streets, which are non-changeable.')
```

#### ВІДПОВІДЬ:
На начальному етапі ми отримали файл с значною кількістю дубльованих та незаповненних значень, що може суттєво спотворити результати дослідження.
Для запобіганння цьому ми прибрали дубльовані значення та заповнили усі NaN значення (як числові, так і категоріальні) відповідними типами даних:
- Числові значення заповнені медіаною усіх інших, заповненних значень у відповідних колонках. Використання медіани дозволяє дещо згладити можливе спотворення при дослідженні, оскільки використання середнього значення буде враховувати і "пусті" дані, тим самим підвищуючи "розкид" значень при дослідженні;
- Для категоріальних значень було використано method=bfill, що дозволяя дотриматись структури та згладити дані.

### 4. Конструювання нових ознак

Цей єтап призначено для поліпшення _'інформативности'_ даних за рахунок додання похідних даних з комбінацій існуючих та/або з зовнішніх джерел

#### 4.1 створити нову колонку `district` використовуючі дані, що знаходяться в файлі 'kijv_districts.csv'

```python
# створити датафрейм `districts` з файлу kijv-districts.csv
districts = pd.read_csv('kijv_districts.csv')
districts = districts.drop_duplicates('street')
districts
```

```python
# написати код, який співвідностить значення показчика `street` обох датасетах та вибирає відповініий вулиці район
apartmentDistricts = apartments.merge(districts[['street', 'district']], on='street')
apartmentDistricts = apartmentDistricts.drop_duplicates()
```

```python
apartmentDistricts
```

#### 4.2 створити нову колонку `area_comfort` на по наступній формулі:

__area_comfort = area_total - area_living - area_kitchen__

_передбачити контроль на позитивне значення результата_

```python
apartmentDistricts['area_comfort'] = (apartmentDistricts['area_total'] - apartmentDistricts['area_living'] - apartmentDistricts['area_kitchen']).clip(lower=0)
```

```python
apartmentDistricts
```

#### 6.1 Аналіз взаємозв'язків показчиків з цільвою змінною

```python
# побудувати гістограму кореляцій показчиків з ціною квартири
pp_data = list(apartmentDistricts.columns[:9]) + ['area_comfort']

plt.bar(apartmentDistricts[pp_data].corr()['Price'][1:].index, apartmentDistricts[pp_data].corr()['Price'][1:], color='green', label='Ціна')
plt.title('Зв\'язок показчиків з ціною')
plt.xticks(rotation=90)
plt.legend()
plt.xlabel('Ознаки')
plt.ylabel('Рівень кореляції')
plt.grid()
```
